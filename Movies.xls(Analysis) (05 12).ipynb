###### 6. Fill missing values in the 'Budget' column with the mean budget.

```python
# Create a new DataFrame 'new_data' to avoid modifying the original DataFrame
new_data = data.copy()

# Calculate the mean budget
mean_budget = new_data['Budget'].mean()

# Fill missing values in the 'Budget' column with the mean budget
new_data['Budget'] = new_data['Budget'].fillna(mean_budget)

# Display the updated DataFrame
print("DataFrame with Missing Values in 'Budget' Filled:")
new_data
```

###### 7. Replace missing values in the 'Language' column with the most common language.


```python
# Create a new DataFrame 'new_lang' to avoid modifying the original DataFrame
data_lang=data.copy()

# Most common language:
Most_common_language=data_lang.Language.value_counts().idxmax()

# Fill missing values in the 'Language' column with the Most common language used in dataset
data_lang['Language'] = data_lang['Language'].fillna(Most_common_language)

# Display the updated DataFrame
print("DataFrame with Missing Values in 'Language' Filled:")
data_lang
```

###### 10. Calculate the percentage of movies with a positive profit (where Gross Earnings > Budget).

```python
# Calculate profit of each movie without modifying the original DataFrame
profit = data['Gross Earnings'] - data['Budget']

# Number of movies with positive profit
positive_profit_count = (profit > 0).sum()   #The sum() function is used in this context because when you perform a comparison operation (profit > 0), it creates a boolean Series, and sum() effectively counts the number of True values in that Series          
#positive_profit_count = len(profit[profit > 0])


# Calculate the percentage of movies with positive profit
print("Percentage of movies with positive profit:", (positive_profit_count / data.shape[0])*100)
```

###### 11. Show top 10 Director from USA who gave maximum returning(profit%) in Comedy|Drama field  (Title , Director, Budget, Gross Earnings , & profit%):-

```python
# Drop rows with NaN values in 'Budget' 'Gross Earnings' 'Director' 'Country' or 'Genres'
data_cleaned = data.dropna(subset=['Budget', 'Gross Earnings','Director','Country','Genres'])

# Calculate 'Profit%' 
data_cleaned['Profit%'] = ((data_cleaned['Gross Earnings'] - data_cleaned['Budget']) / data_cleaned['Budget']) * 100

# Show the top 10 movies by profit percentage
top_10_movies = data_cleaned[(data_cleaned["Genres"]=="Comedy|Drama")&(data_cleaned["Country"]=="USA")].sort_values(by='Profit%', ascending=False).head(10)[['Title','Director','Budget','Gross Earnings','Profit%']]#.reset_index(drop=True)

print("Top 10 Comedy|Drama Movies Director list from USA whose got maximum % of returning:")
top_10_movies
```

###### 12. Identify movies with extremely high user reviews (consider movies with more than 90% of the maximum user reviews).
