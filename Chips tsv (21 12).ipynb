###### 2.Highest Price of chicken item:-

First change dtypes of item_pricecolumn and replace $.

i) User defined 

```python
def convert(x):
    r=x.replace("$"," ")
    r=float(r)
    return r

orders["item_price"]=orders.item_price.apply(convert)
```

ii) Pre defined

orders['item_price'] = orders['item_price'].str.replace('$','').astype(float)
